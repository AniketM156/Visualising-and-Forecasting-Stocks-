### Delete unnecesary columns

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
sp500
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

### Wether the stock price wil go up or down

```python
sp500["Target"] = (sp500["Tomorrow"]>sp500["Close"]).astype(int)
sp500
```

### Removing values before 1990 due to irrelevence

```python
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
# model = RandomForestClassifier(n_estimate=1000,min_sample= (experiment the no.),random_state = 1)
model = RandomForestClassifier(n_estimate=100,min_sample=100,random_state = 1)
# cant use Cross Validation in time series analysis
train = sp500.iloc[:-100] # everything except last 100 values
test = sp500.ioc[-100:] # last 100 values

predictors = ["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```
